## **Sección 1: Extracción, Documentación y Estandarización**

### Diccionario Médico de Datos (Las 10 características base)
Para que usted como Científico de Datos entienda la biología de la que estamos hablando, los doctores midieron 10 aspectos del núcleo de las células. Para cada aspecto calcularon 3 cosas: la media (mean), el error estándar (error) y el peor de los casos (worst), dándonos 30 columnas en total.

- **Radio (Radius):** Distancia del centro a los bordes.
- **Textura (Texture):** Variación en la escala de grises de la imagen celular.
- **Perímetro (Perimeter):** Tamaño del borde del núcleo.
- **Área (Area):** Tamaño total de la célula.
- **Suavidad (Smoothness):** Variación local en la longitud de los radios.
- **Compacidad (Compactness):** (Perímetro^2 / Área) - 1.0
- **Concavidad (Concavity):** Gravedad de las porciones cóncavas del contorno.
- **Puntos Cóncavos (Concave points):** Número de partes hundidas en el contorno.
- **Simetría (Symmetry):** Qué tan simétrica es la célula.
- **Dimensión Fractal (Fractal dimension):** Aproximación de la "rugosidad" del borde.

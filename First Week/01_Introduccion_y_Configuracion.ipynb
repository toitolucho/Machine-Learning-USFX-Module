## **Sección 2: Delimitación de la IA y Definición de Machine Learning**

### **Relación Jerárquica**
*   **Inteligencia Artificial (IA):** Área de las ciencias de la computación orientada al desarrollo de sistemas capaces de emular capacidades cognitivas humanas.
*   **Machine Learning (ML):** Subcampo de la IA enfocado en algoritmos que aprenden patrones directamente a partir de datos estructurados o no estructurados sin reglas de programación explícitas.
*   **Deep Learning (DL):** Subconjunto de ML especializado en el uso de arquitecturas de redes neuronales artificiales profundas.
*   **Ciencia de Datos (Data Science):** Disciplina interdisciplinaria que combina computación, estadística y conocimiento del dominio para extraer información a partir de conjuntos de datos.

```
+-----------------------------------------------------+
| Inteligencia Artificial (IA)                        |
|   +-----------------------------------------------+ |

|   |   +-----------------------------------------+ | |
|   |   | Deep Learning (DL)                      | | |
|   |   +-----------------------------------------+ | |
|   +-----------------------------------------------+ |
+-----------------------------------------------------+
```

### **Definiciones Formales**
*   *Arthur Samuel (1959):* "Campo de estudio que otorga a las computadoras la capacidad de aprender sin ser programadas explícitamente."
*   *Tom Mitchell (1997):* "Un programa de computadora aprende de la experiencia **E** con respecto a una clase de tareas **T** y una medida de rendimiento **P**, si su rendimiento en las tareas en T, medido por P, mejora con la experiencia E."

| Componente | Definición | Aplicación (Detector de Correo no Deseado) |
| :--- | :--- | :--- |
| **Tarea (T)** | Actividad u operación a realizar | Clasificar correos como 'spam' o 'no spam' |
| **Experiencia (E)** | Conjunto de datos históricos de entrada | Correos históricos previamente etiquetados |
| **Rendimiento (P)** | Métrica de evaluación de exactitud | Proporción de correos correctamente clasificados |

### **Comparación del Paradigma de Programación**
*   **Programación Tradicional:** Entrada (Datos + Reglas) $\rightarrow$ Salida (Respuestas).
*   **Machine Learning:** Entrada (Datos + Respuestas) $\rightarrow$ Salida (Reglas/Patrones).

## **Sección 3: Representación de Datos y la Metodología CRISP-DM**

### **Fases del Ciclo de Vida CRISP-DM**
La metodología CRISP-DM (Cross Industry Standard Process for Data Mining) describe el ciclo de vida de un proyecto de análisis de datos a través de seis etapas iterativas:
1. **Comprensión del Negocio (Business Understanding):** Definición de objetivos comerciales e identificación del problema a resolver.
2. **Comprensión de los Datos (Data Understanding):** Recolección inicial, exploración y evaluación de la calidad de los datos.
3. **Preparación de los Datos (Data Preparation):** Limpieza, transformación y estructuración de los datos de entrada para los modelos.
4. **Modelado (Modeling):** Selección y aplicación de técnicas matemáticas y calibración de hiperparámetros.
5. **Evaluación (Evaluation):** Verificación técnica de los resultados y validación con los requisitos del negocio.
6. **Despliegue (Deployment):** Integración del modelo en producción para la toma de decisiones.

### **Estructuración de Datos**
*   **Características (Features) / Matriz $X$:** Variables independientes que describen las observaciones del conjunto de datos.
*   **Etiqueta (Label) / Vector $y$:** Variable dependiente u objetivo que el modelo debe predecir en problemas de aprendizaje supervisado.
*   **Datos Etiquetados:** Registros que contienen tanto variables predictoras como la variable objetivo (usado en Aprendizaje Supervisado).
*   **Datos No Etiquetados:** Registros compuestos únicamente por variables predictoras (usado en Aprendizaje No Investigado).

````python
# Definición de un diccionario con datos clínicos de pacientes
datos_clinicos = {
    'Edad': [25, 47, 52, 36, 22, 58, 40, 31, 45, 60],
    'IMC': [22.4, 28.1, 31.5, 24.2, 19.8, 33.1, 26.5, 23.0, 30.2, 29.5],
    'Presion_Arterial': [120, 130, 140, 115, 110, 145, 125, 118, 135, 138],
    'Diabetico': [0, 1, 1, 0, 0, 1, 0, 0, 1, 1]
}

# Creación del DataFrame de pacientes y visualización directa del objeto
df_pacientes = pd.DataFrame(datos_clinicos)
df_pacientes
````

````python
# Separación en variables independientes (X) y dependiente (y)
X_pacientes = df_pacientes[['Edad', 'IMC', 'Presion_Arterial']]
y_pacientes = df_pacientes['Diabetico']

# Impresión directa de las dimensiones de los conjuntos resultantes
print(X_pacientes.shape)
print(y_pacientes.shape)
````

## **Sección 4: Validación de Modelos y División de Datos**

### **Ajuste de Modelos (Fitting)**
*   **Subajuste (Underfitting):** Ocurre cuando el modelo cuenta con una estructura matemática demasiado simple, impidiéndole capturar el patrón subyacente de los datos de entrenamiento (Alto Sesgo).
*   **Sobreajuste (Overfitting):** Ocurre cuando el modelo memoriza el ruido y los detalles específicos del conjunto de entrenamiento, perdiendo la capacidad de generalizar ante datos nuevos (Alta Varianza).

### **La Regla de Oro de Machine Learning**
Queda establecido que no se debe evaluar el rendimiento de un modelo empleando los mismos datos utilizados en su fase de entrenamiento. Para validar la capacidad de generalización del algoritmo, el conjunto de datos inicial se divide en:
1. **Conjunto de Entrenamiento (Training Set):** Utilizado para que el algoritmo aprenda los parámetros del modelo.
2. **Conjunto de Prueba (Test Set):** Utilizado como conjunto independiente para medir la precisión y generalización del modelo.

````python
from sklearn.model_selection import train_test_split

# Partición de los datos de pacientes (70% entrenamiento y 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(
    X_pacientes, 
    y_pacientes, 
    test_size=0.30, 
    random_state=42
)
````

````python
# Impresión de las dimensiones resultantes de la partición
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
````

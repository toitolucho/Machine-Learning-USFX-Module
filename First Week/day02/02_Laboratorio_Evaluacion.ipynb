```python
# =====================================================================
# BLOQUE DE INICIALIZACIÓN (Ejecutar sin modificar)
# Generación procedimental de los datos del laboratorio
# =====================================================================
import pandas as pd
import numpy as np

np.random.seed(100)
n_viviendas = 150

# Generación de variables continuas y discretas
precios = np.random.normal(loc=150000, scale=40000, size=n_viviendas)
superficies = np.random.normal(loc=120, scale=30, size=n_viviendas)
habitaciones = np.random.choice([2, 3, 4, 5], size=n_viviendas, p=[0.2, 0.5, 0.2, 0.1])
zonas = np.random.choice(['Norte', 'Sur', 'Este', 'Oeste'], size=n_viviendas)

df_inmobiliaria = pd.DataFrame({
    'Precio': precios,
    'Superficie_m2': superficies,
    'Habitaciones': habitaciones,
    'Zona_Ciudad': zonas
})

# Inyección deliberada de ruido y errores de captura
# 1. Inserción de Valores Nulos
idx_nan = np.random.choice(n_viviendas, size=15, replace=False)
df_inmobiliaria.loc[idx_nan, 'Habitaciones'] = np.nan

# 2. Inserción de Outliers (Anomalías Ilógicas)
df_inmobiliaria.loc[5, 'Superficie_m2'] = -20   # Área negativa
df_inmobiliaria.loc[78, 'Superficie_m2'] = 0    # Área inexistente
df_inmobiliaria.loc[120, 'Precio'] = -50000     # Precio negativo

print("¡Base de datos generada e importada al entorno con éxito!")
display(df_inmobiliaria.head(10).style.highlight_null(color='red'))
```
